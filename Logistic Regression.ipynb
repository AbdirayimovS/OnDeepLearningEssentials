```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100,
                          n_features=2,
                          n_informative=2,
                          n_redundant=0, 
                          random_state=42)
```

```python
y = y.reshape(-1, 1)
```

```python
plt.scatter(X[y[:, 0] == 0][:, 0], X[y[:, 0] == 0][:, 1], label="class 0")
plt.scatter(X[y[:, 0] == 1][:, 0], X[y[:, 0] == 1][:, 1], label="class 1")
```

```python
X = np.hstack([np.ones((len(y), 1)), X])
```

```python
learning_rate = 0.01 
epochs = 1000

def sigmoid(z): # z = X@w
    return 1/(1 + np.exp(-z))

def gradient_fn(w, X, y):
    m = len(y) # X.shape[1]
    h = sigmoid(X@w)
    cost = 1/m * X.T @ (h - y)
    return cost 

w = np.zeros((X.shape[1], 1))

for epoch in range(epochs):
    gradient = gradient_fn(w, X, y)
    w = w - learning_rate * gradient
    if epoch % 200 == 0:
        print(f"Epochs: {epoch}\tGradient: {gradient}")
```

```python
w
```

```python
x_values = np.linspace(-1, 1, 100)
y_values = -(w[0] + w[1] * x_values) / w[2] # drawing a line to clearly seperate the classes
```

```python

plt.scatter(X[y[:, 0] == 0][:, 1], X[y[:, 0] == 0][:, 2], label="class 0")
plt.scatter(X[y[:, 0] == 1][:, 1], X[y[:, 0] == 1][:, 2], label="class 1")
plt.plot(x_values, y_values, label='Decision Boundary')
```
